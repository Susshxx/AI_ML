```python
import numpy as np

# Setup Network Parameters
# Inputs: x1 = 0.1, x2 = 0.5 [cite: 14, 19, 186]
x = np.array([[0.1], [0.5]])

# Target Outputs: y1 = 0.05, y2 = 0.95 [cite: 18, 31, 188]
y_target = np.array([[0.05], [0.95]])

# Weights from Input to Hidden Layer (W1) [cite: 190, 202]
# Using values: w5=0.1, w6=0.2, w7=0.3, w8=0.4
W1 = np.array([[0.1, 0.2],
               [0.3, 0.4]])

# Weights from Hidden to Output Layer (W2) [cite: 191, 202]
# Using values: w9=0.5, w10=0.6, w11=0.7, w12=0.8
W2 = np.array([[0.5, 0.6],
               [0.7, 0.8]])

# Biases [cite: 26, 28, 204]
b1 = np.array([[0.25], [0.25]])
b2 = np.array([[0.35], [0.35]])

learning_rate = 0.5
epochs = 500

# Activation Function: Sigmoid [cite: 48, 79, 206]
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivative of Sigmoid for Backpropagation [cite: 50, 83, 215]
def sigmoid_derivative(a):
    return a * (1 - a)

print(f"Starting Training for {epochs} epochs...\n")

# Training Loop
for epoch in range(epochs):

    # FORWARD PASS [cite: 204, 208]
    # Hidden Layer
    z1 = np.dot(W1, x) + b1
    o = sigmoid(z1)

    # Output Layer
    z2 = np.dot(W2, o) + b2
    y = sigmoid(z2)

    # Calculate Total Error (Mean Squared Error) [cite: 44, 74, 105]
    total_error = 0.5 * np.sum((y_target - y)**2)

    # BACKWARD PASS (Computing Gradients) [cite: 211]
    # Error at Output Layer (delta2) [cite: 217]
    # delta = (y - y_target) * sigmoid_derivative
    delta2 = (y - y_target) * sigmoid_derivative(y)

    # Gradient for W2 [cite: 218, 222]
    dW2 = np.dot(delta2, o.T)

    # Error at Hidden Layer (delta1) [cite: 224, 225]
    # delta1 = (W2.T * delta2) * sigmoid_derivative(o)
    delta1 = np.dot(W2.T, delta2) * sigmoid_derivative(o)

    # Gradient for W1 [cite: 226, 228]
    dW1 = np.dot(delta1, x.T)

    # UPDATE WEIGHTS [cite: 234, 236]
    W2 -= learning_rate * dW2
    W1 -= learning_rate * dW1

    # Update Biases (Simplified)
    b2 -= learning_rate * delta2
    b1 -= learning_rate * delta1

    # Print progress every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch} | Error: {total_error:.6f}")

# --- 3. Final Results ---
print("\nTraining Complete:")
print(f"Final Predicted Output:\n{y}")
print(f"Target Output:\n{y_target}")
```

###Summary of the Matrix Operations Used:

1. Forward pass:  In forward pass, we did hidden layer calculation : z1 = np.dot(W1, x) + b1 where, matrix multiplication (W1 × input) + bias has taken place. Similarly, Activation: o = sigmoid(z1) where i appply function to each value (element-wise). Output layer calculation: z2 = np.dot(W2, o) + b2 where there is a matrix multiplication (W2 × hidden output) + bias along with Activation: y = sigmoid(z2)

2. Error Calculation: total_error = 0.5 * np.sum((y_target - y)**2). It represents the element-wise subtraction → square → sum.

3. Backward Pass: Output layer error (delta2) where delta2 = (y - y_target) * sigmoid_derivative(y). It denotes the element-wise multiplication. Then, Gradient for W2: dW2 = np.dot(delta2, o.T). It represents the matrix multiplication (delta × transpose of hidden output)

4. Hidden layer error (delta1): delta1 = np.dot(W2.T, delta2) * sigmoid_derivative(o): It denotes the matrix multiplication + element-wise multiplication. Gradient for W1 represented by dW1 = np.dot(delta1, x.T)

5. Weight Update
W2 -= learning_rate * dW2
W1 -= learning_rate * dW1

6. Bias Update
b2 -= learning_rate * delta2
b1 -= learning_rate * delta1

Neural Network Training ProcessVectorization for Speed:

Instead of updating every single connection one by one with slow loops, we group inputs ($x$) and weights ($W$) into matrices. By using tools like NumPy and the np.dot command, the computer calculates thousands of connections in a single mathematical step, which is significantly more efficient than manual calculations.

The Forward Pass: This is the first stage where the network takes the input ($x$), multiplies it by a weight matrix ($W^{(1)}$), and adds a bias ($b^{(1)}$) to find the pre-activation value ($z^{(1)}$). This value is then passed through a sigmoid function $\sigma(z)$ to create the hidden layer output ($o$), which then repeats the process to generate the final "guess" ($y$) at the output layer.

Calculating the Error: To understand how wrong a guess was, the program calculates the difference between the network's prediction ($y$) and the actual target ($y_{target}$). This difference is combined with the sigmoid derivative to create an Error Signal ($\delta^{(2)}$) for the output layer.

Backpropagation (The Backward Pass): To fix mistakes, the error signal is sent backward through the network. By using a "flipped" version of the weights (the transpose $W^{(2)T}$), the program calculates the error signal for the hidden layer ($\delta^{(1)}$), identifying how much each earlier layer contributed to the total error.
