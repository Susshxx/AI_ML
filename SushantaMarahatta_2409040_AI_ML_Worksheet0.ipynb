**Task 4: Matrix Operations**

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

```python
#1. Prove A · A⁻¹ = I
A_inv = np.linalg.inv(A)

identity = np.dot(A, A_inv)

print("A inverse:\n", A_inv)
print("A * A^-1:\n", identity)
```

```python
#2. Prove AB ≠ BA
AB = np.dot(A,B)
BA = np.dot(B,A)

print("AB:\n", AB)
print("BA:\n", BA)
```

```python
#3. Prove (𝐴𝐵)𝑇=𝐵𝑇𝐴𝑇(AB)T=BTAT
left = np.transpose(np.dot(A,B))
right = np.dot(np.transpose(B), np.transpose(A))

print("Left (AB)^T:\n", left)
print("Right B^T A^T:\n", right)
```

```python
#• Solve the following system of Linear equation using Inverse Methods.

#2x − 3y + z = −1
#x − y + 2z = −3
#3x + y − z = 9

A = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

B = np.array([-1, -3, 9])

# Solve using inverse
A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print("Solution [x, y, z]:")
print(X)
```

```python
import time
size = 1000000
list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result = [list1[i] + list2[i] for i in range(size)]
end = time.time()
print("Python list addition time:", end-start)

#NumPy
arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
result = arr1 + arr2
end = time.time()
print("NumPy addition time:", end-start)
```

```python
#2. Element-wise Multiplication

start = time.time()
result = [list1[i] * list2[i] for i in range(size)]
end = time.time()
print("Python list multiplication time:", end-start)

#NumPy
start = time.time()
result = arr1 * arr2
end = time.time()
print("NumPy multiplication time:", end-start)
```

```python
#3. Dot Product

start = time.time()
dot = sum(list1[i]*list2[i] for i in range(size))
end = time.time()
print("Python dot product time:", end-start)

#NumPy
start = time.time()
dot = np.dot(arr1,arr2)
end = time.time()
print("NumPy dot product time:", end-start)
```

```python
#4. Matrix Multiplication

size = 100
A = [[1]*size for i in range(size)]
B = [[1]*size for i in range(size)]

start = time.time()
result = [[sum(A[i][k]*B[k][j] for k in range(size))
          for j in range(size)] for i in range(size)]
end = time.time()
print("Python matrix multiplication time:", end-start)

#NumPy
A = np.ones((1000,1000))
B = np.ones((1000,1000))
start = time.time()
result = np.dot(A,B)
end = time.time()
print("NumPy matrix multiplication time:", end-start)
```

From the observation, it is clear that NumPy performs much faster than Python lists. In every mathematical operation,
